### Retention и время, которое ученики проводят в приложении

```python
# функция для создания профилей пользователей
def get_profiles(sessions, orders):

    # находим параметры первых посещений
    profiles = (
        sessions.sort_values(by=['user_id', 'open_time'])
        .groupby('user_id')
        .agg(
            {'open_time': 'first'}
        )
        .rename(columns={'open_time': 'first_ts'})
        .reset_index()
    )

    # для когортного анализа определяем дату первого посещения
    # и первый день месяца, в который это посещение произошло
    profiles['dt'] = profiles['first_ts'].dt.date
    profiles['month'] = profiles['first_ts'].astype('datetime64[M]')

    # добавляем признак платящих пользователей
    profiles['payer'] = profiles['user_id'].isin(orders['user_id'].unique())

    # считаем количество уникальных пользователей
    # с одинаковой датой начала активности
    new_users = (
        profiles.groupby(['dt'])
        .agg({'user_id': 'nunique'})
        .rename(columns={'user_id': 'unique_users'})
        .reset_index()
    )

    return profiles
```

```python
# получение профилей пользователей
profiles = get_profiles(sessions, payments)
display(profiles.head())
```

#### Retention

```python
# объединим profiles и users
profiles = profiles.sort_values(by='user_id')
profiles_users = (
    profiles.merge(users, on='user_id', how='left')
    .sort_values(by='reg_time')
)
profiles_users.head()
```

```python
# определим минимальную и максимальную даты активности пользователей
min_date = profiles_users['dt'].min()
max_date = profiles_users['dt'].max()
print('Минимальная дата начала активности пользователей:', min_date)
print('Максимальная дата начала активности пользователей:', max_date)
```

```python
# задаём момент и горизонт анализа данных
# на календаре 16 июня 2023, и зададим горизонт анализа
observation_date = datetime(2023, 6, 8).date()
horizon_days = 10
```

```python
# функция для расчёта удержания

def get_retention(
    profiles,
    sessions,
    observation_date,
    horizon_days,
    dimensions=[],
    ignore_horizon=False,
):

    # добавляем столбец payer в передаваемый dimensions список
    dimensions = ['payer'] + dimensions

    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')

    # собираем «сырые» данные для расчёта удержания
    result_raw = result_raw.merge(
        sessions[['user_id', 'open_time']], on='user_id', how='left'
    )
    result_raw['lifetime'] = (
        result_raw['open_time'] - result_raw['first_ts']
    ).dt.days

    # функция для группировки таблицы по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days):
        result = df.pivot_table(
            index=dims, columns='lifetime', values='user_id', aggfunc='nunique'
        )
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        result = result.div(result['cohort_size'], axis=0)
        result = result[['cohort_size'] + list(range(horizon_days))]
        result['cohort_size'] = cohort_sizes
        return result

    # получаем таблицу удержания
    result_grouped = group_by_dimensions(result_raw, dimensions, horizon_days)

    # получаем таблицу динамики удержания
    result_in_time = group_by_dimensions(
        result_raw, dimensions + ['dt'], horizon_days
    )

    # возвращаем обе таблицы и сырые данные
    return result_raw, result_grouped, result_in_time
```

```python
# функция для визуализации удержания
def plot_retention(retention, retention_history, horizon, window=7):

    # задаём размер сетки для графиков
    plt.figure(figsize=(15, 10))

    # исключаем размеры когорт и удержание первого дня
    retention = retention.drop(columns=['cohort_size', 0])
    # в таблице динамики оставляем только нужный лайфтайм
    retention_history = retention_history.drop(columns=['cohort_size'])[
        [horizon - 1]
    ]

    # если в индексах таблицы удержания только payer,
    # добавляем второй признак — cohort
    if retention.index.nlevels == 1:
        retention['cohort'] = 'All users'
        retention = retention.reset_index().set_index(['cohort', 'payer'])

    # в первой строим кривые удержания платящих пользователей
    ax1 = plt.subplot(2, 2, 1)
    retention.query('payer == True').droplevel('payer').T.plot(
        grid=True, ax=ax1
    )
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('Удержание платящих пользователей')

    # во второй ячейке строим кривые удержания неплатящих
    # вертикальная ось — от графика из первой ячейки
    ax2 = plt.subplot(2, 2, 2, sharey=ax1)
    retention.query('payer == False').droplevel('payer').T.plot(
        grid=True, ax=ax2
    )
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('Удержание неплатящих пользователей')

    # в третьей ячейке — динамика удержания платящих
    ax3 = plt.subplot(2, 2, 3)
    # получаем названия столбцов для сводной таблицы
    columns = [
        name
        for name in retention_history.index.names
        if name not in ['dt', 'payer']
    ]
    # фильтруем данные и строим график
    filtered_data = retention_history.query('payer == True').pivot_table(
        index='dt', columns=columns, values=horizon - 1, aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax3)
    plt.xlabel('Дата привлечения')
    plt.title(
        'Динамика удержания платящих пользователей на {}-й день'.format(
            horizon
        )
    )

    # в четвёртой ячейке — динамика удержания неплатящих
    ax4 = plt.subplot(2, 2, 4, sharey=ax3)
    # фильтруем данные и строим график
    filtered_data = retention_history.query('payer == False').pivot_table(
        index='dt', columns=columns, values=horizon - 1, aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax4)
    plt.xlabel('Дата привлечения')
    plt.title(
        'Динамика удержания неплатящих пользователей на {}-й день'.format(
            horizon
        )
    )

    plt.tight_layout()
    plt.show()
```

```python
# функция для сглаживания фрейма
def filter_data(df, window):
    # для каждого столбца применяем скользящее среднее
    for column in df.columns.values:
        df[column] = df[column].rolling(window).mean()
    return df
```

```python
retention_raw, retention, retention_history = get_retention(
    profiles, sessions, observation_date, horizon_days, dimensions = [], ignore_horizon=False
)
plot_retention(retention, retention_history, horizon_days)
```

```python
plt.figure(figsize = (15, 5))
sns.heatmap(retention.drop(columns=['cohort_size']),
            annot=True,
            fmt='.2%',
            vmin=0,
            vmax=1,
            center=0.8
           )
plt.title('Тепловая карта удержания пользователей')
plt.show()
```

**Вывод**  
Удержание платящих пользователей выше, нежели чем у неплатящих пользователей на 10 день удержания.

#### Время, которое ученики проводят в приложении

```python
# продолжительность сессии в минутах
sessions['duration'] = ((sessions['close_time'] - sessions['open_time']) / np.timedelta64(1, 'm')).round(2)
display(sessions.head())
```

```python
# суммарная продолжительность сессий, общее количество сессий, средняя продолжительность сессии
sessions_calc = (
    sessions.pivot_table(
    index = 'user_id',
    values = 'duration',
    aggfunc = {'sum', 'count'}
    )
    .rename(columns={'sum': 'total_duration', 'count': 'number_ses'})
    .sort_values(by='user_id')
    .reset_index()
)

sessions_calc['avg'] = sessions_calc['total_duration'] / sessions_calc['number_ses']
display(sessions_calc.head())
```

```python
# среднее количество сессий (медианное значение)
median_number = round(sessions_calc['number_ses'].median())
print('Среднее количество сессий:', median_number)
```

```python
# средняя продолжительность сессии на каждого пользователя
fig = plt.figure(figsize = (15, 5))

ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

y = sessions_calc['avg']
ax1.boxplot(y)
ax1.set_xlabel('Средняя продолжительность сессии, мин.')
ax2.boxplot(y)
ax2.set_xlabel('Средняя продолжительность сессии, мин.')
ax2.set_ylim([0, 20])
```

**Вывод**  
На основании графиков, можно сделать вывод, что существует большое количество выбросов, поэтому необходимо использование медианы, а не среднего значения.

```python
# средняя продолжительность сессий (медианное значение)
median_ses = round(sessions_calc['avg'].median())
print('Средняя продолжительность сессий:', median_ses, 'мин.')
```

**Общий вывод**  
1. На 10 день платящие пользователи удерживаются лучше неплатящих.  
2. Среднее количество сессий: 2.  
3. Средняя продолжительность сессий: 9 мин.  

### Как ученики переходят с уровня на уровень

```python
levels_users = (
    users.merge(levels, on = 'user_id', how = 'left')
    .sort_values(by=['user_id', 'time'])
)
display(levels_users.head())
```

```python
# распределение играков по уровням

users_per_levels = (
    levels_users.pivot_table(
        index = 'level',
        values = 'user_id',
        aggfunc = 'nunique'
        )
    .rename(columns={'user_id': 'total_count'})
    .sort_values(by='total_count', ascending = False)
    .reset_index()
)

display(users_per_levels.head())
```

```python
# функция для столбчатой диаграммы aquamarine
def bar(data, x, y, text, title, xaxis_title, yaxis_title):
    fig = px.bar(data,
                 x=x,
                 y=y,
                 text=text,
                 color_discrete_sequence=px.colors.qualitative.Pastel2
                 )
    fig.update_layout(title=title,
                      xaxis_title=xaxis_title,
                      yaxis_title=yaxis_title
                     )
    fig.show()
```

```python
x_name = 'level'
y_name = 'total_count'
text = 'total_count'
title_name = 'Количество пользователей на уровнях'
xaxis_title_name = 'Уровень'
yaxis_title_name = 'Количество пользователей'

bar(users_per_levels,
    x_name,
    y_name,
    text,
    title_name,
    xaxis_title_name,
    yaxis_title_name
   )
```

```python
# сколько времени ученики проводят на уровнях
levels_users = levels_users.dropna()
```

```python
levels_users['diff'] = levels_users.groupby('user_id')['time'].diff()
display(levels_users.head())
```

```python
users_levels_dur = levels_users.dropna()
display(users_levels_dur.head())
```

```python
users_levels_dur['duration_level'] = (
    users_levels_dur['diff'].dt.total_seconds()/60)\
    .round(2)

display(users_levels_dur.head())
```

```python
# распределение времени между уровнями

plt.figure(figsize=(20, 10))
sns.boxplot(data=users_levels_dur, x='level', y='duration_level')
plt.xlabel('Новый уровень пользователя')
plt.ylabel('Время прохождения уровня в мин.')
plt.title('Распределение времени между уровнями')
plt.show()
```

**Вывод**  
На основании графика можно сделать вывод, что разброс времени очень большой. Возможно, что некоторые пользователи проходят уровни с перерывом.

```python
median_duration = (
    users_levels_dur.groupby('level')['duration_level']
    .median()
    .reset_index()
)
```

```python
x_name = 'level'
y_name = 'duration_level'
text = 'duration_level'
title_name = 'Средняя продолжительность перехода с уровня на уровень'
xaxis_title_name = 'Уровень'
yaxis_title_name = 'Продолжительность в мин.'

bar(median_duration,
    x_name,
    y_name,
    text,
    title_name,
    xaxis_title_name,
    yaxis_title_name
   )
```

**Вывод**  
Первые 7 уровней пользователи проходят с легкостью, быстро. Однако продолжительность перехода с 8 на 9, с 9 на 10 уровни резко возрастает. Наиболее долгий переход наблюдается с уровня 11 на уровень 12.
